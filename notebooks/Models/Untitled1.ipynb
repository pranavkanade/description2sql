```python
import json

data_path_mapping = '/root/projects/description2sql/data/dev.jsonl'
data_path_tables = '/root/projects/description2sql/data/dev.tables.jsonl'
synthesized_dataset = './dev_demo.json'

def get_data(path):
    with open(path, 'r') as fp:
        json_data = (list(map(json.loads, fp.readlines())))
    return json_data
```

```python
q_list = [
    "What is the date of entry for the UK Albums Top 75 chart?",
    "What was the total number of weeks on peak for the Ireland Albums Top 75 chart",
    "What is the exit date for the Dutch Albums Top 100 Chart?",
    "What is the title of the episode with the production code 624?",
    "what's the species in the world with peruvian amazon vs. peru (percent)  of 63",
    "what's the minimum species in the peruvian amazon with species in peru of 1000",
    "what's the maximum peru vs. world (percent) with 9672 species in the world",
    "What school did the forward whose number is 10 belong to?",
    "What was the evening gown score when the swimsuit was 9.61?",
    "What is the swimsuit score when the interview was 9.74?",
    "Which country had an interview score of 9.40 and average of 9.44?",
    "Which country had the swimsuit score 9.67?",
    "What was the average for the country with the swimsuit score of 9.57?",
    "What is the date of Game 50?",
    "Who scored the most points in Game 49?",
    "What is the team located at philips arena 18,227?",
    "what are all the records with a score is w 98–91",
    "who are all the visitor with a record of 25–18",
    "Which visitors have a leading scorer of roy : 25",
    "what is the total number of dates where the scorer is gordon : 32",
    "How many times was the # of total votes 2582322?",
    "What year was the election when the # of seats won was 65?",
    "What is the election year when the # of candidates nominated was 262?",
    "Who is the top scorer where gf is 41?",
    "How many goals were scored in the 2005-06 season?",
    "How many seasons had a rank of 16?",
    "Who is the manager whose rank is 16?",
    "What was the result of the Top 3 episode?"
]
```

```python
tab_list = list()
test_mapping = list()
```

```python
mapping = get_data(data_path_mapping)
```

```python
tabs = get_data(data_path_tables)
```

```python
mapping[0]
```

```python
for each_q in q_list:
    for each_map in mapping:
        if each_q == each_map['question']:
            tab_list.append(each_map['table_id'])
            test_mapping.append(each_map)
```

```python
tab_list = set(tab_list)
```

```python
len(tab_list)
```

```python
tab_list = list(tab_list)
```

```python
test_tables = list()
```

```python
for tab_id in tab_list:
    for each_tab in tabs:
        if tab_id == each_tab['id']:
            test_tables.append(each_tab)
```

```python
len(test_tables)
```

```python
for idx in range(len(test_tables)):
    test_tables[idx]['questions'] = [m for m in test_mapping if m['table_id'] == test_tables[idx]['id']]
```
